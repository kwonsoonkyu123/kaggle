```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
import seaborn as sns
%matplotlib inline
```

```python
with open("datasets/train.json", encoding = "UTF8") as data_file:
    train_data = json.load(data_file)
with open("datasets/test.json", encoding = "UTF8") as data_file:
    test_data = json.load(data_file)
train_data = pd.DataFrame(train_data)
test_data = pd.DataFrame(test_data)
```

```python
# cuisine들의 countplot
g = sns.countplot("cuisine", data = train_data)
for label in g.get_xticklabels():
    label.set_rotation(90)
```

```python
# ingredients의 list를 multi row로 변환
lens = [len(ingredient) for ingredient in train_data["ingredients"]]
train_data_for_graph = pd.DataFrame( {"cuisine" : np.repeat(train_data["cuisine"].values,lens), 
               "id" : np.repeat(train_data["id"].values,lens),
               "ingredient" : np.hstack(train_data["ingredients"])})
```

```python
import itertools
# 전체 ingredients의 count plot

# figure 생성
fig, axes = plt.subplots(5, 2, figsize=(15,20))

axes = list(itertools.chain.from_iterable(axes))

# top 10의 ingredient
top_10 = train_data_for_graph.ingredient.value_counts()[0:10].index
for ingredient, ax, idx in zip(top_10, axes, range(10)):
    sns.countplot(x = "ingredient", hue = "cuisine",
                  data = train_data_for_graph[train_data_for_graph.ingredient == ingredient], ax = ax,
                  hue_order = train_data_for_graph[train_data_for_graph.ingredient == ingredient].cuisine.value_counts().index)
    if (idx % 4 < 2):
        ax.legend(loc = "center right", framealpha = 0.2)
    else :
        ax.legend(loc = "center left", framealpha = 0.1)
#     ax.legend_.remove()
# handles, labels = ax.get_legend_handles_labels()
# fig.legend(handles, labels, loc="right")
```

```python
# 나라별 top 10 ingredient plot
figures, axes = plt.subplots(5, 4, figsize=(15,20))
axes = list(itertools.chain.from_iterable(axes))

cuisines = train_data_for_graph.cuisine.value_counts().index

for cuisine, ax in zip(cuisines, axes):
    ingredients = train_data_for_graph[train_data_for_graph.cuisine == cuisine].ingredient.value_counts().index[0:9]
    sns.countplot(y = "ingredient", ax = ax, order = ingredients,
                  data = train_data_for_graph[(train_data_for_graph.cuisine == cuisine) &
                                              (train_data_for_graph.ingredient.isin(ingredients))])
    ax.set(ylabel = "", title = cuisine, xlabel = "")
    




# top_ingredient_by_cuisine = pd.DataFrame(train_data_for_graph.groupby(["cuisine"]).ingredient.value_counts().groupby(["cuisine"]).head())
# cuisines = top_ingredient_by_cuisine.index.get_level_values("cuisine").value_counts().index

# for cuisine, ax in zip(cuisines, axes):
#     sns.barplot(x = top_ingredient_by_cuisine.index.get_level_values("ingredient"), y = "ingredient", ax = ax,
#                   data = top_ingredient_by_cuisine.iloc[top_ingredient_by_cuisine.index.get_level_values("cuisine") == cuisine])
```
